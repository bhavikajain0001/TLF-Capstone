```python
df_groundtruth_test = pd.read_csv('/content/drive/MyDrive/projects/Capstone_Project/Checkpoint_5/GAN-BERT/Q2/test_Q2.tsv',sep='\t')
df_pred_test = pd.read_csv('/content/drive/MyDrive/projects/Capstone_Project/Checkpoint_5/GAN-BERT/Q2/ganbert_output_model/predict_Q2/test_results.tsv',sep='\t',header=None)

df_groundtruth_val = pd.read_csv('/content/drive/MyDrive/projects/Capstone_Project/Checkpoint_5/GAN-BERT/Q2/valid_Q2.tsv',sep='\t')
df_pred_val = pd.read_csv('/content/drive/MyDrive/projects/Capstone_Project/Checkpoint_5/GAN-BERT/Q2/ganbert_output_model/predict_Q2/val_results.tsv',sep='\t',header=None)
```

```python
df_groundtruth_test['sub_intent'] = df_groundtruth_test['fine_label utterance'].apply(lambda x: x.split(" ")[0])
df_groundtruth_val['sub_intent'] = df_groundtruth_val['fine_label utterance'].apply(lambda x: x.split(" ")[0])
```

```python
# This label list is printed during the inference process. You can directly copy the list from there
label_list = ['UNK_UNK', 'Explore_params', 'Effect_ofparams', 'Dependencies_bwparams','Selected_params', 'Monitoring_objectives','Tradeoff_bw_objectives']
df_pred_test.columns = label_list 
df_pred_val.columns = label_list 
```

```python
df_pred_test['sub_intent'] = df_pred_test.apply(lambda x: np.argmax(x),axis=1).apply(lambda x: label_list[x])
df_pred_val['sub_intent'] = df_pred_val.apply(lambda x: np.argmax(x),axis=1).apply(lambda x: label_list[x])
```
